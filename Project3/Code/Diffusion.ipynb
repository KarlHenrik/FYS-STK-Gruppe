# The diffusion equation

```python
import numpy as np
import matplotlib.pyplot as plt

from diffusionExplicit import expScheme
from diffusionTf import DiffusionTf
from diffusionUtils import *
```

## dx = 0.1

```python
dx = 0.1
L = 1
dt = dx**2 * 0.5
T = 0.5
```

```python
model_01 = DiffusionTf(dx, dt, L, T)
x, t, U = model_01.train(10000)
error(x, t, U)
plot3D(x, t, U, 'Neural network')
plot2D(x, t, U, 'Neural network')
```

```python
x, t, U = expScheme(dx, dt, L, T)
error(x, t, U)
plot3D(x, t, U, 'Explicit Scheme')
plot2D(x, t, U, 'Explicit Scheme')
```

```python
x, t, U = analyticSolv(dx, dt, L, T)
error(x, t, U)
plot3D(x, t, U, 'Analytical solution')
plot2D(x, t, U, 'Analytical solution')
```

### The code crashes when the stability criterion for the explicit scheme is neglected:

```python
dx = 0.01
L = 1

dt = dx**2 * 0.51 #dt is here more than dx^2 *0.5 which is the stability criterion
T = 0.5
```

```python
x, t, U = expScheme(dx, dt, L, T)
error(x, t, U)
plot3D(x, t, U, 'Explicit Scheme with dt > dx^2 * 0.5 and')
plot2D(x, t, U, 'Explicit Scheme with dt > dx^2 * 0.5 and')
```
